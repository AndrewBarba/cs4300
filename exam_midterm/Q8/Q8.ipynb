```python
# Dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from skimage import color
from skimage import img_as_float, img_as_ubyte, img_as_uint
import glob

# Set defaults
plt.rcParams['image.cmap'] = 'gray'
plt.rcParams['image.interpolation'] = 'none'
plt.rcParams['figure.figsize'] = 14, 14

# Import images
imgpaths = glob.glob('./images/*.jpg')
imgset = [img_as_ubyte(mpimg.imread(x)) for x in imgpaths]

# Display thumbnails of the images to ensure loading
plt.figure()
for i,img in enumerate(imgset):
    plt.subplot(1, len(imgset), i+1)
    plt.imshow(img, cmap = 'gray')
```

```python
# easy plotting
def plot(result):
    plt.figure()
    plt.subplot(1, 2, 1)
    plt.imshow(result[0], cmap = 'gray')
    plt.subplot(1, 2, 2)
    plt.imshow(result[1], cmap = 'gray')
```

```python
from scipy import signal as sg

# Kernals
HORIZANTAL_KERNAL = [[[1]], 
                   [[-1]]]

VERTICAL_KERNAL = [[[1, -1]]]

def find_edges(img):
    img = img.copy()
    h = sg.convolve(img, HORIZANTAL_KERNAL)
    v = sg.convolve(img, VERTICAL_KERNAL)
    return [v, h]
```

```python
for img in imgset:
    r = find_edges(img)
    plot(r)
```

Conclusion
======

I chose to use scipy for this question because of it's robust API for working with arrays. Afterall, an image is just a multidimensional array of colors (pixels) and therefore made this question pretty easy to work with. The one "gotcha" in this question was making sure the kernal was same the dimension as the image. We can't use a typical 1 x 2 or 2 x 1 kernal because in order to properly convolve the 2 matrices they need to be the same dimension. Knowing this, I adjusted our vertical and horizantal kernal becore using scipy's built in convolve method.
